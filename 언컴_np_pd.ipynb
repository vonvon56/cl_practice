Numpy

```python
import numpy as np
import re
```

```python
a = [[1,2,3],[4,5,6]]
#np 형식으로 만들기
a_np = np.array(a)

print(a)
print(a_np)
print(a_np.sum())
```

```python
#full
a = np.full(6,5)
print(a)
a = np.full([3,4],1)
print(a)
a = np.random.random([3,4])
print(a)
```

```python
#배열 생성하는 다른 방법 - arange
np_10 = np.arange(2,10)
print(np_10)
#np.zeros(), np.ones()
np_zero = np.zeros((3,4))
print(np_zero)
#n행m열로 변환하기: .reshape()
reshaped = np_10.reshape(2,4)
print(reshaped)
```

```python
#배열에서 특정 원소의 인덱스 찾기: where
spot = str(np.where(reshaped == 7))
print(spot)
x,y = map(int,re.findall(r'(?<=\[)[0-9]+(?=\])', spot))
print(x,y)
print(reshaped[x,y:])
```

```python
#배열 연산
arr = np.arange(15).reshape(3,5) + 1
print(arr)
print(np.max(arr, axis=1)) #1: 각 행에서 최대
print(np.max(arr, axis=0)) #0: 각 열에서 최대

print(np.mean(arr, axis=0))

print(np.mean(arr, axis=1))
```

```python
#조건 주면 T/F 담아서 array로 리턴
arr > 9
print(arr)
#True인 곳 찾기 : np.where()
boolean = arr>9
print(boolean)
haha = np.where(boolean)
haha
```

```python
import numpy as np

arr = np.array([[1, 5, 12, 8],
                [9, 3, 7, 10],
                [15, 2, 6, 11]])



# np.where()를 사용하여 True인 위치를 찾습니다.
haha = np.where(arr>9)

# haha를 출력합니다.
print(haha)
arr[haha]
```

```python
#표준편차, 분산
array1 = np.array([14, 6, 13, 21, 23, 31, 9, 5])

print(array1.std()) # 표준 편차
print(array1.var()) # 분산
```

```python
#중앙값
array1 = np.array([8, 12, 9, 15, 16])
array2 = np.array([14, 6, 13, 21, 23, 31, 9, 5])

print(np.median(array1)) # 중앙값
print(np.median(array2)) # 중앙값
```

```python
#평균값
array1 = np.array([14, 6, 13, 21, 23, 31, 9, 5])

print(array1.mean()) # 평균값
```

```python
####index_col='name'으로 맨 처음 column명 지을 수도 있다.
names = ['grade','class','name','message']
students = pd.read_csv('students.csv', header =None, names=names, index_col='name')
students
```

```python
names = ['grade','class','name','message']
students = pd.read_csv('students.tsv', header =None, names=names, sep='\t')
print(students)
students.loc[3, 'message'] = '''aaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaa
aaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaa
aaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaa'''
#loc: 인덱스 슬라이싱. 1:3이면 1부터 2까지 아니고 3까지.
#iloc: 상대적인 위치로 접근. 이건 진짜 index처럼 다룸. 1:3 -> 2번째부터 3번째

print(students.iloc[1:3])
students
```

```python
#대표적 통계치 출력: describe()
students.describe()
```

```python
#정렬 : axis = 0: 행 기준
students.sort_index(axis=0,  ascending = False)
```

```python
#정렬: axis=1: 열 기준
students.sort_index(axis=1)
```

```python
#groupby: 묶기
students.groupby('grade').mean()
```

```python
#apply() : 여러 cell에 대해 일괄적으로 하나의 함수 적용
def add_1(grade):
    if grade == 3:
        new_grade = 0
    else:
        new_grade = grade + 1
    return new_grade

students['grade'] = students['grade'].apply(add_1)
students
```
